### Sigmoid function Python implementation

Sigmoid function is special case of logistic function that we have seen in earlier notebook

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x_axis = np.linspace(-10, 10, 100)
y_axis = np.array([sigmoid(i) for i in x_axis])
```

```python
plt.plot(x_axis, y_axis)
plt.grid(True)
```

Derivative of sigmoid function

```python
def sigmoid_derivative(x):
    return sigmoid(x)*sigmoid(-x)
```

```python
y_axis = np.array([sigmoid_derivative(i) for i in x_axis])
```

```python
plt.plot(x_axis, y_axis)
plt.grid(True)
```
