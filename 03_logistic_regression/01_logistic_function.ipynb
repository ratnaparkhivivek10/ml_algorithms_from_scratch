### Logistic function from scratch

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

This is original logistic function used to study population growth.

```python
# e = the natural logarithm base
# x0 = the x-value of the sigmoid's midpoint
# L = the curve's maximum value
# k = the steepness of the curve
# the graph of f approaching L as x approaches +∞ and approaching zero as x approaches −∞ and the function has midpoint=x0
# and slope=k

def logistic_function(x, x0, L, k):
    return L / (1 + np.exp(-k*(x-x0)))
```

Lets plot this function with different values of x0, L and k

```python
x_axis = np.linspace(-10, 10, 100)
```

midpoint(x0) = 0
<br>
max value of curve(L) = 1
<br>
slope/steepneess of the curve = 1
<br>

This will become sigmoid function used in logistic regression

```python
y_axis = np.array([logistic_function(i, x0=0, L=1, k=1) for i in x_axis])
```

```python
plt.plot(x_axis, y_axis)
plt.grid(True)
```

Observe the above graph and see the function has slope=1, max value=1 and mid point=0

Let's change the value again
<br>
midpoint(x0) = 2.5
<br>
max value of curve(L) = 5 
<br>
slope/steepneess of the curve = 0.5 
<br>

```python
y_axis = np.array([logistic_function(i, x0=2.5, L=5, k=0.5) for i in x_axis])
```

```python
plt.plot(x_axis, y_axis)
plt.grid(True)
```

Let's change the values again

```python
y_axis = np.array([logistic_function(i, x0=0, L=1, k=2) for i in x_axis])
```

```python
plt.plot(x_axis, y_axis)
plt.grid(True)
```

You can try the different combinations of maxvalue(L), midpoint(x0) and slope(k) and plot the function. In logistic regression we need the probability of the class, the probability values lies between 0 and 1, so, we have to modify this function to take the maximum value 1, slope 1 and midpoint 0.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```
